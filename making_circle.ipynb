```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
fig = plt.gcf()
ax = fig.add_subplot(111,aspect='equal')
ax.scatter(0,0,s=40,color='black',alpha=0.9)
ax.plot(np.linspace(0,1,100),np.zeros(100),color='black')
#np.linspace(0,1,100)
#np.zeros(100)
ax.set_xlim(left=-1,right=1)
ax.set_ylim(bottom=-1,top=1)
ax.add_artist(plt.Circle((0,0),1,alpha=0.5))
ax.text(0.4,0.1,'r',size=30)
```

```python
def make_circle(point=0):
    fig = plt.gcf()
    ax = plt.subplot(111,aspect='equal')
    ax.scatter(0,0,s=30,color='black')
    ax.plot(np.linspace(0,1,100),np.zeros(100),color='black')
    ax.set_xlim(left=-1,right=1)
    ax.set_ylim(top=1,bottom=-1)
    ax.add_artist(plt.Circle((0,0),1,alpha=0.5))
    ax.text(0.4,0.1,'r',color='black',alpha=0.9,size=30)
    plt.xlabel('circle')
    plt.ylabel('not a corcle')
    plt.title('circle',size=30,alpha=0.5,color='red')
    
    if point:
        ax.text(0.55,0.9,'far away',alpha=0.7,size=10,color='black')
        ax.scatter(0.85,0.85,color='black',alpha=0.5,s=20)
    else:
        plt.show()
        
make_circle()
        
```

```python
make_circle(2)
```
